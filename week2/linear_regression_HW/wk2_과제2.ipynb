# 나머지 변수들은 상관관계계수를 bar형태로 표현해 보았다.

```python
numerical_features
```

```python
label = list()
y_value = list()
for feature in numerical_features:
    try:
        print(data[feature].corr(data['Hammer_price']))
        label.append(feature)
        y_value.append(data[feature].corr(data['Hammer_price']))
    except:
        print(feature, "는", type(feature), "이기 때문에 형변환이 필요합니다.")
```

```python
index = np.arange(len(label))
plt.figure(figsize=(30,5))
plt.title('Numerical 데이터와 Hammer price간의 상관계수들')
plt.bar(index, y_value, tick_label=label, align='center')
```

### price쪽은 hammer_price와 엄청난 양의 상관관계를 가지는 것으로 보이고 결국 낙찰가는 총감정가나 최저가격에서 큰 차이가 없는 것을 볼 수 있다.
### 상관계수가-0.5 ~ 0.5 사이의 값은 큰 의미가 없다.

# 다중공선성 확인

```python
category_feature = [ col for col in data.columns if data[col].dtypes == "object"]
numerical_feature = list(set(data.columns) - set(category_feature))
print(category_feature)
print(numerical_feature)
```

```python
sns.pairplot(data[numerical_feature])
```

```python
vif_data = data[numerical_feature].copy()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])]
vif["features"] = data.columns
vif
```
