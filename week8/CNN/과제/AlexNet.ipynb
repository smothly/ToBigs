```python
import torch.nn as nn
import torch.utils.model_zoo as model_zoo


__all__ = ['AlexNet', 'alexnet']


model_urls = {
    'alexnet': 'https://download.pytorch.org/models/alexnet-owt-4df8aa71.pth',
}


class AlexNet(nn.Module):

    def __init__(self, num_classes=1000):
        super(AlexNet, self).__init__()
        # fully connected layer 전까지의 
        # convolution 과정 디테일하게 relu까지 전부 추가하였습니다.
        # 기존에 input output size나 채널이 안맞은거 수정하였음.
        self.features = nn.Sequential(
            # layer1 
            nn.Conv2d(3, 96, kernel_size=11, stride=4, padding=0),
            nn.ReLU(inplace=True),
            # layer2
            nn.MaxPool2d(kernel_size=3, stride=2),
            # layer3
            nn.Conv2d(96, 256, kernel_size=5, stride=1, padding=2),
            nn.ReLU(inplace=True),
            # layer4
            nn.MaxPool2d(kernel_size=3, stride=2),
            # layer5
            nn.Conv2d(256, 384, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            # layer6
            nn.Conv2d(384, 384, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            # layer7
            nn.Conv2d(384, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            # layer8
            nn.MaxPool2d(kernel_size=3, stride=2),
        )
        # fullu connected 부분
        self.classifier = nn.Sequential(
            # layer9 & flatten
            nn.Linear(6 * 6 * 256, 4096),
            nn.ReLU(inplace=True),
            # layer 10
            nn.Linear(4096, 4096),
            nn.ReLU(inplace=True),
            # layer 11 & softmax
            nn.Linear(4096, num_classes)
#             torch.nn.Softmax(dim),
        )

    def forward(self, x):
        x = self.features(x)
        # flatten
        print(x.shape)
        x = x.view(-1, 6 * 6 * 256)
        x = self.classifier(x)
        return x

# 출처 : https://pytorch.org/docs/0.4.0/_modules/torchvision/models/alexnet.htm
# ? 채운 모델과 달라 약간 수정하고 주석 추가 하였습니다!
```
